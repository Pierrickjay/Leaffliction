```python
%matplotlib widget
from plantcv import plantcv as pcv
from plantcv.parallel import WorkflowInputs
```

```python
args = WorkflowInputs(
    images=["/mnt/nfs/homes/mbocquel/PostCommonCore42/Leaffliction/leaves/images/Apple_Black_rot/image (1).JPG"],
    names="image1",
    result="gray_results.json",
    outdir=".",
    writeimg=True,
    debug="plot"
    )

# Set variables
```

```python
img, filename, path = pcv.readimage(filename=args.image1, mode="")
```

```python
pattern1_mask = pcv.threshold.binary(gray_img=img_hsv, threshold=30, object_type='dark')
```
